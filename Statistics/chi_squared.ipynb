```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import pandas as pd

plt.style.use('dark_background')
```

```python
raw_data = pd.DataFrame(np.random.choice([0, 1], size=(100, 2)), columns=['A', 'B'])
df = pd.crosstab(raw_data['A'], raw_data['B'])  # contingency table
df['total'] = df.sum(axis=1)
df.loc['total'] = df.sum()

df
```

### Teste de associação

```python
a = 0.05
chi2_stat, p_val, dof, ex = stats.chi2_contingency(df.iloc[:-1, :-1])
print(f"Chi2 Stat: {chi2_stat}, P-value: {p_val}, Degrees of Freedom: {dof}")
if p_val < a:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

### Teste de aderência

```python
esp = np.array([25, 25, 25, 25])
chi2_stat, p_val = stats.chisquare(df.iloc[:-1, :-1].values.flatten(), f_exp=esp)
print(f"Chi2 Stat: {chi2_stat}, P-value: {p_val}")
if p_val < a:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

### Associação 1xn

```python
contingency_df = pd.DataFrame({
    'Basic': [100, 120],
    'Default': [150, 100],
    'Premium': [50, 80],
})
contingency_df.loc['total'] = contingency_df.sum()
contingency_df['total'] = contingency_df.sum(axis=1)

contingency_df.index = ['Male', 'Female', 'total']
contingency_df
```

```python
chi2_stat, p_val, dof, ex = stats.chi2_contingency(contingency_df.iloc[:-1, :-1])
print(f"Chi2 Stat: {chi2_stat}, P-value: {p_val}, Degrees of Freedom: {dof}")
if p_val < a:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```
