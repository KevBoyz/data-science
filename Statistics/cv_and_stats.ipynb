```python
import pandas as pd
from statsmodels.stats.weightstats import DescrStatsW

func1, func2 = pd.DataFrame(), pd.DataFrame()
```

```python
_range = list(range(2, 7+1))

func1['p'] = [0.1, 0.1, 0.3, 0.2, 0.2, 0.1]
func1['t'] = _range

func2['p'] = [0.08, 0.02, 0.3, 0.2, 0.3, 0.1]
func2['t'] = _range
```

```python
stats1 = DescrStatsW(func1['t'], func1['p'])
stats2 = DescrStatsW(func2['t'], func2['p'])

mean1, mean2 = stats1.mean, stats2.mean
std1, std2 = stats1.std, stats2.std

cv1, cv2 = std1 / mean1, std2 / mean2
```

```python
print(mean1, std1, cv1)
print(mean2, std2, cv2)
```
