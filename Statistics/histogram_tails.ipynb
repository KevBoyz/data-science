```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def tail(dados):
    sns.set_theme(style="whitegrid")
    plt.figure(figsize=(10, 6))
    plot = sns.histplot(dados, kde=True, color="skyblue", bins=30)
    plt.axvline(np.mean(dados), color='red', linestyle='--', label=f'Média: {np.mean(dados):.2f}')
    plt.axvline(np.median(dados), color='green', linestyle='-', label=f'Mediana: {np.median(dados):.2f}')
    plt.title('Histograma com Assimetria Positiva (Cauda à Direita)', fontsize=15)
    plt.xlabel('Valores')
    plt.ylabel('Frequência')
    plt.legend()
    plt.show()
```

```python
from math import log10, sqrt 
dados = np.random.lognormal(mean=0, sigma=0.7, size=1000)
```

```python
tail(dados)
```

```python
tail([log10(d) for d in dados])
```

```python
tail([sqrt(d) for d in dados])
```

```python
tail([1/d for d in dados])
```

```python
p90 = np.percentile(dados, 95)
winson = [d if d <= p90 else p90 for d in dados]
```

```python
tail(winson)
```
