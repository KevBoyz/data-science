```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# 1. Gerar amostras fictícias (Médias e Variâncias diferentes)
np.random.seed(42)
grupo1 = np.random.normal(loc=25, scale=2, size=80)   # Média ~25, Variância ~4
grupo2 = np.random.normal(loc=21, scale=4, size=50)   # Média ~21, Variância ~16

# 2. Executar o Teste t de Welch (equal_var=False)
# H1: Média do Grupo 1 > Média do Grupo 2
t_stat, p_val = stats.ttest_ind(grupo1, grupo2, equal_var=False, alternative='greater')

# 3. Calcular Magnitude (d de Cohen)
d_cohen = (np.mean(grupo1) - np.mean(grupo2)) / np.sqrt((np.std(grupo1)**2 + np.std(grupo2)**2) / 2)

# 4. Gerar o Histograma Combinado
plt.figure(figsize=(10, 6))
plt.hist(grupo1, bins=8, alpha=0.5, label=f'Fertilizante A (Média: {np.mean(grupo1):.1f})', color='blue')
plt.hist(grupo2, bins=8, alpha=0.5, label=f'Fertilizante B (Média: {np.mean(grupo2):.1f})', color='orange')

# Adicionar linhas das médias para visualizar o "Sinal"
plt.axvline(np.mean(grupo1), color='blue', linestyle='dashed', linewidth=2)
plt.axvline(np.mean(grupo2), color='orange', linestyle='dashed', linewidth=2)

plt.title(f'Teste t de Welch\nEstatística t: {t_stat:.3f} | Valor-p: {p_val < 0.05} | d de Cohen: {d_cohen:.2f}')
plt.xlabel('Crescimento (cm)')
plt.ylabel('Frequência')
plt.legend()
plt.show()
```
