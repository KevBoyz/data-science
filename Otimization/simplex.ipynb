```python
"""
max z = 2x1 + x2

sujeito a

x1 + 2x2 <= 6
2x1 + x2 <= 6
x1, x2 >=0
"""

import numpy as np

# matriz associada a x1 e x2 da inequacao
A = np.array([[1, 2],
              [2, 1]])

# matrix limitante da inequacao b1 b2
b = np.array([[6], [6]])

# c associado a funcao objetivo 
c = np.array([[2, 1]])

# A primeira tabela
# base |x,  s, b
#---------------
# s    |A,  I, b
# z    |-c, 0, 0

m, _ = np.shape(A)
I = np.eye(m)

# Parte de cima da tabela
T11 = np.concatenate((A, I, b), axis=1)
# Parte de baixo da tabela
T21 = np.concatenate((-c, np.zeros((1, m+1))), axis=1)

# Tabela Inicial: montagem da tabela
T0 = np.concatenate((T11, T21), axis=0)

# Achar Quem entra
e = np.argmin(T0[-1, :])

# Quem sai
Razao = T0[0:-1, -1]/T0[0:-1, e]
# Achar quem sai
s = np.argmin(Razao)

# Manipulacoes de Pivoteamento
# Transformar o Pivo em 1
T0[s, :] = T0[s, :]/T0[s, e]

# Isso aqui e a L1 = L1 - L2
T0[0, :] = T0[0, :] - T0[s, :]

# Isso aqui: L3 = L3 + 2*L2
T0[-1, :] = T0[-1, :] + 2*T0[s, :]
```
